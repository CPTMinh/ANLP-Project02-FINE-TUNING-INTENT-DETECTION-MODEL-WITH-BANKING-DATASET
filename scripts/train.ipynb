### 5. Update Config Paths to Drive
This modifies your `train.yaml` so the Python script knows to save outputs directly to your Google Drive folder.

```python
import yaml
import os

# Define the Google Drive backup path
drive_path = '/content/drive/MyDrive/ANLP-Intent-Model-Backup'
os.makedirs(drive_path, exist_ok=True)

# Load existing configuration
with open('configs/train.yaml', 'r') as f:
    config = yaml.safe_load(f)

# Update paths to point to Google Drive
config['output_dir'] = f"{drive_path}/outputs"
config['save_model_path'] = f"{drive_path}/saved_model"

# Save the updated config back to train.yaml
with open('configs/train.yaml', 'w') as f:
    yaml.dump(config, f)

print(f"Configurations updated! Checkpoints and final model will be saved live to:")
print(f" - Outputs: {config['output_dir']}")
print(f" - Final Model: {config['save_model_path']}")
```

### 7. Run Inference & Evaluation
Now that the model is trained and saved, we can instantly test it using our inference scripts.

Look for model path

```python
import yaml
import os

drive_saved_model_path = '/content/drive/MyDrive/ANLP-Intent-Model-Backup/saved_model'

# Ensure the path exists
if not os.path.exists(drive_saved_model_path):
    print(f"Warning: Cannot find path {drive_saved_model_path} - check your Google Drive!")

# Update the inference config
with open('configs/inference.yaml', 'r') as f:
    infer_config = yaml.safe_load(f)

infer_config['save_model_path'] = drive_saved_model_path

with open('configs/inference.yaml', 'w') as f:
    yaml.dump(infer_config, f)

print(f"Inference config updated! Testing model from: {drive_saved_model_path}")
```
